```python
%sql
CREATE SCHEMA IF NOT EXISTS workspace.salesstore;
```

```python
%sql
CREATE OR REPLACE TABLE workspace.salesstore.bronze_compras (
  venta_id STRING,
  factura STRING,
  fecha_orden STRING,
  fecha_entrega STRING,
  fecha_envio STRING,
  estado STRING,
  cliente_code STRING,
  nombres STRING,
  apellidos STRING,
  departamento STRING,
  metodo_pago STRING,
  fecha_importacion STRING,
  fecha_carga TIMESTAMP)
USING delta
LOCATION 'abfss://datalake@saproyectoventas.dfs.core.windows.net/bronze/compras'
```

```python
%sql
CREATE OR REPLACE TABLE workspace.salesstore.bronze_detalles (
  detalle_id STRING,
  factura STRING,
  categoria STRING,
  subcategoria STRING,
  producto STRING,
  unidades STRING,
  precio_unitario STRING,
  nombre_archivo STRING,
  fecha_importacion STRING,
  fecha_carga TIMESTAMP)
USING delta
LOCATION 'abfss://datalake@saproyectoventas.dfs.core.windows.net/bronze/detalles'
```

```python
%sql
CREATE OR REPLACE TABLE workspace.salesstore.silver_compras (
  periodo STRING,
  venta_id INT,
  factura STRING,
  fecha_orden DATE,
  fecha_entrega DATE,
  fecha_envio DATE,
  estado STRING,
  cliente_id INT,
  num_documento STRING,
  nombre_cliente STRING,
  departamento STRING,
  metodo_pago STRING,
  dias_envio INT,
  fecha_carga TIMESTAMP)
USING delta
LOCATION 'abfss://datalake@saproyectoventas.dfs.core.windows.net/silver/compras'
PARTITIONED BY (periodo)
```

```python
%sql
CREATE OR REPLACE TABLE workspace.salesstore.silver_detalles (
  detalle_id INT,
  factura STRING,
  tienda STRING,
  categoria STRING,
  subcategoria STRING,
  producto STRING,
  precio_unitario DOUBLE,
  unidades INT,
  subtotal DOUBLE,
  fecha_carga TIMESTAMP)
USING delta
LOCATION 'abfss://datalake@saproyectoventas.dfs.core.windows.net/silver/detalles'
```

```python
%sql
CREATE OR REPLACE TABLE workspace.salesstore.gold_fact_compras (
  periodo STRING,
  factura STRING,
  venta_id INT,
  fecha_orden DATE,
  fecha_envio DATE,
  dias_envio INT,
  estado STRING,
  metodo_pago STRING,
  cliente_id INT,
  num_documento STRING,
  nombre_cliente STRING,
  departamento STRING,
  detalle_id INT,
  tienda STRING,
  categoria STRING,
  subcategoria STRING,
  producto STRING,
  unidades INT,
  subtotal DOUBLE,
  fecha_carga TIMESTAMP)
USING delta
LOCATION 'abfss://datalake@saproyectoventas.dfs.core.windows.net/gold/fact-compras'
PARTITIONED BY (periodo)
```
